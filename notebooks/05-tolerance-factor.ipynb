```python
import pandas as pd
import numpy as np

a_sites_monovalent = {
    "AM": 146,
    "HY": 140,
    "MAM": 216,
    "MA": 217,
    "HZ": 220,
    "HA": 216,
    "PH": 212,
    "AZ": 238,
    "FA": 253,
    "EA": 274,
    "FO": 258,
    "DMA": 272,
    "MH": 275,
    "AZE": 250,
    "HMA": 275,
    "CA": 293,
    "TMA": 294,
    "PY": 278,
    "TTMA": 292,
}

a_sites_divalent = {
    "HZ": 217,
    "AMZ": 267
}

b_sites = {
    "Zr": 72,
    "Hf": 71,
    "Sb": 60,
    "Bi": 76,
    # "Sn": 69,
    # "Pb": 77.5,
}

x_sites = {
    "S": 184,
    "Se": 196,
}

records = []
for b_site, b_r in b_sites.items():
    for x_site, x_r in x_sites.items():
        if b_site in ["Sb", "Bi"]:
            for a_site, a_r in a_sites_monovalent.items():
                records.append((a_site, b_site, x_site, a_r, b_r, x_r))
        else:
            for a_site, a_r in a_sites_divalent.items():
                records.append((a_site, b_site, x_site, a_r, b_r, x_r))

df = pd.DataFrame(records, columns=["A_site", "B_site", "X_site", "A_radius", "B_radius", "X_radius"])
df["tolerance_factor"] = (df["A_radius"] + df["X_radius"]) / (np.sqrt(2) * (df["B_radius"] + df["X_radius"]))
df["octahedral_factor"] = df["B_radius"] / df["X_radius"]
df["compound"] = df["A_site"] + df["B_site"] + df["X_site"] + "$_3$"
```

```python
# mask = df.B_site == "Bi"
# latex = df[mask].copy()[["compound", "A_radius", "B_radius", "X_radius", "tolerance_factor", "octahedral_factor"]]
latex = df.copy()[["compound", "A_radius", "B_radius", "X_radius", "tolerance_factor", "octahedral_factor"]]
latex.rename(columns={"compound": "Compound", "A_radius": "$R_A$ (pm)", "B_radius": "$R_B$ (pm)", "X_radius": "$R_X$ (pm)", "tolerance_factor": "$t$", "octahedral_factor": "$\mu$"}, inplace=True)
print(latex.to_latex(index=False, escape=False,formatters={"$t$": "{:.2f}".format, "$\mu$": "{:.2f}".format}))
```
